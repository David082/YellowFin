### adam

```python
dir_list_adam = [general_dir + "/adam_lr_1.0",
           general_dir + "/adam_lr_0.1",
           general_dir + "/adam_lr_0.01",
           general_dir + "/adam_lr_0.001",
           general_dir + "/adam_lr_0.0001"]
dir_list_mom = [general_dir + "/sgd_mom_lr_1.0",
           general_dir + "/sgd_mom_lr_0.1",
           general_dir + "/sgd_mom_lr_0.01",
           general_dir + "/sgd_mom_lr_0.001"]
dir_list_ours = [general_dir + "/resnet-meta-bundle-acc"]
```

### for 1.01 value

```python
opt_factors = [1.01, 1.05, 1.1, 1.5, 2, 4]
# opt_factors = [2, ]
```

```python
steps_adam=dict()
steps_mom=dict()
steps_ours=dict()
for opt_factor in opt_factors:
    best_step_adam, _ = get_best_steps_to_target_loss(dir_list_adam, win_width, optimal * opt_factor)
    steps_adam[opt_factor] = best_step_adam
    print "adam ", best_step_adam

    best_step_mom, _ = get_best_steps_to_target_loss(dir_list_mom, win_width, optimal * opt_factor)
    steps_mom[opt_factor] = best_step_mom
    print "mom ", best_step_mom

    best_step_ours, _ = get_best_steps_to_target_loss(dir_list_ours, win_width, optimal * opt_factor)
    steps_ours[opt_factor] = best_step_ours
    print "ours ", best_step_ours
```

```python
print "adam steps ", steps_adam
print "mom steps ", steps_mom
print "ours steps", steps_ours
```

```python
speed_up_adam = dict()
speed_up_mom = dict()
speed_up_ours = dict()
# for key in steps_ours.keys():
#     if steps_adam[key] != None:
#         speed_up_adam[key] = steps_adam[key] / float(steps_ours[key] )
#     else:
#         speed_up_adam[key] = None
#     if steps_mom[key] != None:
#         speed_up_mom[key] = steps_mom[key] / float(steps_ours[key] )
#     else:
#         speed_up_mom[key] = None
# print "adam speed up ", speed_up_adam
# print "mom speed up ", speed_up_mom
for key in steps_ours.keys():
    if steps_adam[key] != None:
        speed_up_adam[key] = steps_adam[key] / float(steps_ours[key] )
    else:
        speed_up_adam[key] = None
    if steps_mom[key] != None:
        speed_up_mom[key] = steps_mom[key] / float(steps_ours[key] )
    else:
        speed_up_mom[key] = None
print "adam speed up ", speed_up_adam
print "mom speed up ", speed_up_mom
```

### about learning rate

```python
dir = "../resnet/resnet_res/resnet-meta-bundle-acc"
```

```python
# file_names = ["/lr_0.txt", "/lr_1.txt", "/lr_2.txt", "/lr_3.txt"]
file_names = []
for i in range(18):
    file_names.append("/lr_" + str(i) + ".txt")
lr_max = []
lr_min = []
lr_arg_max = []
lr_arg_min = []
for file_name in file_names:
    f = dir + file_name
    l = np.loadtxt(f).tolist()
    lr_max.append(np.max(l) )
    lr_min.append(np.min(l) )
    
    lr_arg_max.append(np.argmax(l) )
    lr_arg_min.append(np.argmin(l) )
    
print "ResNet max lr (each bundle) ", lr_max
print "ResNet min lr (each bundle) ", lr_min

print "ResNet max lr (each bundle) ", lr_arg_max
print "ResNet min lr (each bundle) ", lr_arg_min


print "Resnet Ratio ", np.array(lr_max) / np.array(lr_min)
```
