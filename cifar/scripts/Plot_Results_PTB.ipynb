```python
dir_list = [general_dir + "/meta-bundle"]
win_width = 500
```

```python
optimal = np.min(best_loss_bundle)
print optimal
```

```python
dir_list_adam = [general_dir + "/adam_lr_1.0",
           general_dir + "/adam_lr_0.1",
           general_dir + "/adam_lr_0.01",
           general_dir + "/adam_lr_0.001",
           general_dir + "/adam_lr_0.0001"]
dir_list_mom = [general_dir + "/sgd_lr_1.0_mom_0",
           general_dir + "/sgd_lr_0.1_mom_0",
           general_dir + "/sgd_lr_0.01_mom_0",
           general_dir + "/sgd_lr_0.001_mom_0",
           general_dir + "/sgd_lr_1.0_mom_0.9",
           general_dir + "/sgd_lr_0.1_mom_0.9",
           general_dir + "/sgd_lr_0.01_mom_0.9",
           general_dir + "/sgd_lr_0.001_mom_0.9",]
dir_list_ours = [general_dir + "/meta-bundle"]
```

```python
opt_factors = [1.01, 1.05, 1.1, 1.5]
# opt_factors = [2, ]
```

```python
def get_loss(dir, win_width):
  if os.path.isfile(dir + "/loss_full.txt"):
    file_name = dir + "/loss_full.txt"
  elif os.path.isfile(dir + "/loss.txt"):
    file_name = dir + "/loss.txt"
  else:
    print dir
    raise Exception("loss file not found")

  loss = np.loadtxt(file_name)
  loss = running_mean(loss, win_width)
  return loss


def get_best_steps_to_target_loss(dir_list, win_width, target_loss):
  steps = dict()
  best_step = None
  for dir in dir_list:
    loss = get_loss(dir, win_width)
    if np.min(loss) > target_loss:
        steps[dir] = None
    else:
        steps[dir] = np.argmax(loss < target_loss)
        
    if best_step == None:
        best_step = steps[dir]
    elif steps[dir] != None and best_step > steps[dir]:
        best_step = steps[dir]
  return best_step, steps
```

```python
steps_adam=dict()
steps_mom=dict()
steps_ours=dict()
for opt_factor in opt_factors:
    best_step_adam, _ = get_best_steps_to_target_loss(dir_list_adam, win_width, optimal * opt_factor)
    steps_adam[opt_factor] = best_step_adam
    print "adam ", best_step_adam

    best_step_mom, _ = get_best_steps_to_target_loss(dir_list_mom, win_width, optimal * opt_factor)
    steps_mom[opt_factor] = best_step_mom
    print "mom ", best_step_mom

    best_step_ours, _ = get_best_steps_to_target_loss(dir_list_ours, win_width, optimal * opt_factor)
    steps_ours[opt_factor] = best_step_ours
    print "ours ", best_step_ours
```

```python
print "adam steps ", steps_adam
print "mom steps ", steps_mom
print "our steps ", steps_ours
```
